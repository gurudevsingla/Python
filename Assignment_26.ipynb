### Question 1 : What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

#### Matplotlib is a powerful data visualization library for Python that is used to create static, animated, and interactive visualizations in Python. It provides a wide range of plotting options and tools to create a variety of high-quality visualizations.
#### Matplotlib is a cross-platform, data visualization and graphical plotting library for Python and its numerical extension NumPy. As such, it offers a viable open source alternative to MATLAB. Developers can also use matplotlibâ€™s APIs (Application Programming Interfaces) to embed plots in GUI applications.
#### Matplotlib is used for several purposes such as:
Creating line plots, scatter plots, bar plots, and histograms to explore and visualize data

Creating publication-quality figures and plots

Generating visualizations for reports and presentations

Creating custom visualizations with unique specifications and designs

Visualizing machine learning models and results

#### The Pyplot module of Matplotlib is a collection of functions that provides a MATLAB-like interface for creating visualizations in Python. Here are five plots that can be created using the Pyplot module:
Line Plot: A line plot is a graph that displays data as a series of points connected by straight lines. This type of plot is useful for visualizing trends and changes over time.

Scatter Plot: A scatter plot is a graph that displays the relationship between two variables by plotting individual data points. This type of plot is useful for visualizing correlations and outliers.

Bar Plot: A bar plot is a graph that displays data as rectangular bars, where the length of each bar corresponds to the value of the variable being plotted. This type of plot is useful for comparing the values of different categories.

Histogram: A histogram is a graph that displays the distribution of a dataset by dividing it into a set of bins and counting the number of observations in each bin. This type of plot is useful for understanding the shape of the data and identifying outliers.

Pie Chart: A pie chart is a circular graph that displays the relative size of each category as a slice of the pie. This type of plot is useful for showing proportions and percentages.




### Question 2 : What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

### Note: Also add title, xlabel, and ylabel to the plot.

```python
from matplotlib import pyplot as plt
plt.scatter(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('ScatterPlot')
plt.show()
```

### Question 3 : Why is the subplot() function used? Draw four line plots using the subplot() function
### Use the following data:

```python
import numpy as np

# Line 1
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

#### The subplot() function in Matplotlib is used to create multiple plots in the same figure.
#### This function allows us to divide the figure into a grid of rows and columns, and then specify which subplot to plot in each cell of the grid.
#### The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the current subplot.

```python
# Importing Matplotlib
import matplotlib.pyplot as plt

# Create four line plots in one figure using subplot()
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

# Plot 1 : Line 1
axes[0, 0].plot(x1, y1)
axes[0, 0].set_title('Line 1')

# Plot 2 : Line 2
axes[0, 1].plot(x2, y2)
axes[0, 1].set_title('Line 2')

# Plot 3 : Line 3
axes[1, 0].plot(x3, y3)
axes[1, 0].set_title('Line 3')

# Plot 4 : Line 4
axes[1, 1].plot(x4, y4)
axes[1, 1].set_title('Line 4')

# Add a common x-axis label and y-axis label
fig.text(0.5, 0.04, 'X-axis', ha='center')
fig.text(0.04, 0.5, 'Y-axis', va='center', rotation='vertical')

# Display the plot
plt.show()
```

### Question 2 : What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

#### Bar graphs are the pictorial representation of data (generally grouped), in the form of vertical or horizontal rectangular bars, where the length of bars are proportional to the measure of data. They are also known as bar charts. Bar graphs are one of the means of data handling in statistics.
#### In a bar plot, the height or length of each bar represents the value or frequency of the corresponding category.
#### Here are some specific reasons why bar plots are used:
1.**Comparing quantities** : Bar plots are used to compare the quantities of different categories, such as sales figures, revenue, or counts of items. By visually comparing the heights or lengths of the bars, it's easy to see which category has the largest or smallest value.

2.**Tracking changes over time** : Bar plots can also be used to track changes over time within a single category. For example, you might create a bar plot showing the revenue of a single department for each month of the year. By comparing the heights or lengths of the bars for each month, you can see how the revenue changes over time.

3.**Showing relationships** : Bar plots can also be used to show relationships between categories. For example, you might create a stacked bar plot showing the sales figures for different products in different regions. By stacking the bars on top of each other, you can see how the sales figures for each product compare across different regions.

4.**Communicating results** : Bar plots are a simple and effective way to communicate results to a non-technical audience. By presenting data in a visually appealing format, bar plots can help to convey the main points of your analysis quickly and clearly.

#### Overall, bar plots are a useful tool for exploring and presenting categorical data in a way that is easy to understand and interpret.

```python
# Plotting Bar plot
from matplotlib import pyplot as plt
plt.bar(company,profit)
plt.xlabel('Company Name')
plt.ylabel('Profit')
plt.title('Profit Comparison for various Companies')
plt.show()
```

```python
# Plotting Horizontal Bar plot
import matplotlib.pyplot as plt
plt.barh(company,profit)
plt.xlabel('Profit')
plt.ylabel('Company Name')
plt.title('Profit Comparison for various Companies')
plt.show()
```

### Question 5 : What is a box plot? Why is it used? Using the following data plot a box plot

```python
import numpy as np

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
```

#### A box plot is a graphical representation of the distribution of a dataset based on five summary statistics, often called the "five-number summary". The five-number summary consists of:
1.**Minimum value** : The smallest observation in the dataset.

2.**First quartile (Q1)** : The value that separates the lowest 25% of the dataset from the rest.

3.**Median (Q2)** : The value that separates the lowest 50% of the dataset from the highest 50%.

4.**Third quartile (Q3)** : The value that separates the lowest 75% of the dataset from the rest.

5.**Maximum value** : The largest observation in the dataset.
A box plot consists of a rectangular box that spans from the first quartile (Q1) to the third quartile (Q3), with a vertical line at the median (Q2). The "whiskers" extend from the top and bottom of the box to the highest and lowest values that are still within 1.5 times the interquartile range (IQR) of the box. Any points that fall outside of the whiskers are considered outliers and are plotted as individual points.

#### Box plots are commonly used to visualize the distribution of a dataset and to identify potential outliers or extreme values. They are particularly useful for:
**Summarizing the distribution of a dataset** : Box plots provide a visual summary of the distribution of a dataset based on its five-number summary (minimum, first quartile, median, third quartile, and maximum). This allows us to quickly understand the range, central tendency, and spread of the data.

**Identifying potential outliers** : Box plots can also be used to identify potential outliers or extreme values in a dataset. Any data points that fall outside of the whiskers (i.e., more than 1.5 times the interquartile range from the nearest quartile) are considered outliers and are plotted as individual points. This allows us to quickly identify any unusual or extreme values in the data.

**Comparing distributions** : Box plots can also be used to compare the distribution of two or more datasets. By plotting multiple box plots side by side, we can compare the range, central tendency, and spread of each dataset and identify any differences or similarities between them.

#### Overall, box plots are a powerful tool for visualizing and summarizing the distribution of a dataset, and are widely used in data analysis, statistical modeling, and data visualization.
Plotting boxplots for example mentioned in question

```python
# Importing Matplotlib
import matplotlib.pyplot as plt

# Combine the data into a list
data = [box1, box2]

# Create the box plot
fig, ax = plt.subplots()
ax.boxplot(data,patch_artist=True, boxprops=dict(facecolor='lightblue'))

# Add labels and title
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Values')
ax.set_title('Box Plot of Two Datasets')

# Display the plot
plt.show()
```
