### Question 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot

#### Five plots in Seaborn along with their usage is as following :
1.**Box Plot** : A box plot is a graphical representation of data that shows the distribution of a dataset by indicating the median, quartiles, and outliers. It is used to identify the spread and skewness of the data.

2.**Heatmap** : A heatmap is a graphical representation of data that uses color-coding to represent the values of a two-dimensional dataset. It is used to identify patterns in the data, such as correlations or clusters.

3.**Pair Plot** : A pair plot is a graphical representation of data that shows pairwise relationships between different variables in a dataset. It is used to identify correlations between variables and to explore the relationships between 
them.

4.**Violin Plot** : A violin plot is a graphical representation of data that combines elements of a box plot and a kernel density plot. It is used to visualize the distribution of a dataset and show its density.

5.**Joint Plot** : A joint plot is a graphical representation of data that shows the relationship between two variables using a scatter plot and a histogram. It is used to explore the relationship between two variables and to identify correlations between them.
#### These are just a few examples of the types of plots that can be created using Seaborn. Seaborn offers many other visualizations that can be used to explore and analyze different types of data.

```python
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('fmri')
print(df.head())
plt.figure(figsize=(8,4))
sns.lineplot(x='timepoint',y='signal',data=df,style='region',hue='event')
plt.title('Signal over Time for Different Events and Regions')
plt.show()
```

### Question 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.
#### Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset("titanic")
print(df.head())
sns.boxplot(x='pclass',y='age',data=df)
plt.title("pclass vs age")
```

```python
sns.boxplot(x='pclass',y='fare',data=df)
plt.title("pclass vs fare")
```

### Question 4 : Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
df=sns.load_dataset('diamonds')
print(df.head())
sns.histplot(x='price',hue='cut',data=df,kde=True)
plt.title('Distribution of Price for various diamond cut')
plt.show()
```

### Question 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset

```python
import seaborn as sns
df=sns.load_dataset('iris')
print(df.head())
sns.pairplot(hue='species',data=df)
```

### Question 6 : Use the "flights" dataset from seaborn to plot a heatmap

```python
import seaborn as sns
flights=sns.load_dataset('flights')
flights.head()
```

```python
flights_matrix = flights.pivot(index="year", columns="month", values="passengers")
flights_matrix
```

```python
import matplotlib.pyplot as plt
sns.heatmap(flights_matrix, 
            cmap="YlGnBu", 
            annot=True,
            fmt="d",
            linewidths=.5,
            linecolor="black")
plt.title("Passengers per Month")
plt.show()
```
