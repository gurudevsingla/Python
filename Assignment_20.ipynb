#### Q1. Create a Pandas Series that contains the following data: 4, 8, 15, 16, 23, and 42. Then, print the series.
#### Ans 1.

```python
import pandas as pd
l = [4, 8, 15, 16, 23, 42]
i = [1,2,3,4,5,6] # index 
s = pd.Series(l, index = i)
print(s)
```

#### Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series?
#### Ans 4.
A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

DataFrame is different from pandas series because DataFrame is Mutable, 2 dimensional and DataFrame objects can have different data types while pandas series is immutable, one dimensional and all elements must be of same data type.

#### Q7. Create a DataFrame using multiple Series. Explain with an example.
#### Ans 7.

```python
import pandas as pd

# Created a series "s" with column name "Subjects"
s = pd.Series(["Python", "Pandas", "NumPy", "Data Science"], name = "Subjects")

# Created another series "s1" with column name "Grades"
s1 = pd.Series(["A+", "A", "B", "B+"], name = "Grades")

# Merging the s and s1
df1 = pd.merge(s,s1,right_index = True,left_index = True)
df1
```
